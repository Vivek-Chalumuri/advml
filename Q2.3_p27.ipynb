```python
import numpy as np
import pickle
import pandas as pd
from ex_2_3 import Node, load_params, load_sample, print_tree
from ex_2_3_tree_helper import Tree
```

```python
my_data_path = ''

with open(my_data_path + '../2_3/tree_params.pickle', 'rb') as handle:
    params = pickle.load(handle, encoding='latin1')

with open(my_data_path + '../2_3/tree_samples.pickle', 'rb') as handle:
    samples = pickle.load(handle, encoding='latin1')
```

```python
def PofBeta(root):
    result = 0
    #iterate for all possible values of the sample
    for i in range(K):
        result += sub_problem(root, i)*root.cat[0][i]
    return result

def sub_problem(node, i):
    #check if leaf node
    name = str(node)+'_'+ str(i)
    if name in sd_dict.keys():
        #return from saved values if previously evaluated
        return sd_dict[name]
    
    if len(node.descendants) == 0:
        if node.sample == i:
            sd_dict[name] = 1
            return 1
        else:
            sd_dict[name] = 0
            return 0
    else:
        P = 1
        #iterate over descendants of the node
        for child in node.descendants:
            S = 0
            #iterate for all possible values of the sample
            for j in range(K):
                S += child.cat[i][j]*sub_problem(child, j)
            P *= S
        #save the value in a dictionary for reuse
        sd_dict[name] = P
        return P
```

```python
df = pd.DataFrame(columns=['Sample_Name', 'P'])
j = 0
for t in range(1,4):
    for i in range(len(list(params.keys()))):
        params_name = list(params.keys())[i]     
        params_l = params[params_name]
        root = load_params(params_l)
        K = len(root.cat[0])
        sd_dict = {}
        sname = '_sample_'+str(t)
        samples_name = params_name + sname
        sample = samples[samples_name]
        load_sample(root, sample)
        result = PofBeta(root)
        print(samples_name)
        print(result)
        df.loc[j] = [samples_name, result]
        j += 1
        print(j,"_"*50)
```
